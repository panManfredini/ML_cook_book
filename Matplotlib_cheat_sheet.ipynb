```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

# Subplot Syntax

```python
# Subplots syntax
columns = 2
rows = 1

fig, (axis1,axis2) = plt.subplots(rows, columns, figsize=(10,5))

axis1.plot(np.arange(10),np.sin(np.arange(10)))
axis2.plot(np.arange(10),np.sin(np.arange(10)))
```
